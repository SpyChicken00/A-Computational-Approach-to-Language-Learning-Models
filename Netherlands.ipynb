```python
import pandas as pd
netherlands = pd.read_csv('/content/drive/My Drive/stex.csv', encoding='ISO-8859-1')
print(netherlands.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sample = netherlands.sample(10000)
scatter = sns.scatterplot(data=netherlands, x="Enroll", y="Speaking", hue = "L1")
scatter.set_position([3, 3, 2, 2])
# Get the current legend and move it to the desired location
#plt.legend(loc='upper left')

# Show the plot
plt.show()
```

```python
jp_sample = sample[sample["L1"] == "Japanese"]
sns.scatterplot(data=jp_sample, x="Enroll", y="Speaking", hue = "L1")
```
